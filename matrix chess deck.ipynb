```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import random 
from  random import randint 
from math import exp
```

```python
while True:
    n = int(input("Введите число больше 2-х: "))
    if n>2:
        break

a = 0
b = 0
mat = []
randomlist = [1, -1]
mat1 = []
mat2 = []
T = 0.8

while a < n:
    while b < n:
        mat1.append(random.choice(randomlist))
        b += 1
    mat2.append(mat1)
    b = 0
    mat1 = []
    a += 1
    
mat = np.array(mat2)
print(mat)
def mtx(x, y=0):
    for i in range(n): 
        for j in range(n-1):
            y += x[i][j] * x[i][j+1]
    for i in range(n-1):
        for j in range(n):
            y += x[i][j] * x[i+1][j]
        
    y += x[0][0]*x[0][n-1]
    y += x[0][0]*x[n-1][0]
    y += x[n-1][0]*x[n-1][n-1]
    y += x[0][n-1]*x[n-1][n-1]

    for i in range(n-1):
        if i == 0:
            continue
        if i == n-1:
            continue
        y += x[n-1-i][0]*x[n-1-i][n-1]

    for i in range(n-1):
        if i == 0:
            continue
        if i == n-1:
            continue
        y += x[0][n-1-i]*x[n-1][n-1-i]
    return y

for item in range(501):

    sum = mtx(mat)
    print('Сумма новой конфигурации равна', sum)
    matt = mat.copy()
    
    c = random.randint(0, n-1) # Задание №2
    d = random.randint(0, n-1)
    print('изменяется ячейка', '[', c, ']', '[', d, ']')
    matt[c][d] = matt[c][d] * -1
    print('E2=')
    print(matt)

    summ = mtx(matt)
    print('Сумма временной конфигурации равна', summ)

    delt = summ - sum # step 4
    print('delt=', delt)
    if delt <= 0:
        mat = matt.copy()
        print('Возвращаемся к шагу 2 и принимаем новую конфигурацию')
    else:
        print('Переходим к шагу 5')
        delt = -delt/T # step 5
        w = np.exp(delt)
        print('w=', w)
        P = random.randint(0, 1000000000000000000)
        P = P/1000000000000000000
        print('P=', P)
        if P <= w:
            mat = matt.copy()
            print('Возвращаемся к шагу 2 меняя конфигурацию')
        else:
            print('Возвращаемся к шагу 2 не меняя конфигурацию')
```
